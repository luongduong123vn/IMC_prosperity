```python
import numpy as np
import pandas as pd
df_constipated = pd.read_csv('./round-4-island-data-bottle/day_1_consolidated.csv')
df_constipated.head()
```

```python
def calc_basket_orchids_return(df, window, beta):
    df[f'basket_return_{window}its'] = (df['GIFT_BASKET'] - df['GIFT_BASKET'].shift(500))/df['GIFT_BASKET'].shift(500)
    df[f'expected_orchids_return_{window}its'] = beta * df[f'basket_return_{window}its']
    df[f'actual_orchids_return_{window}its'] = (df['ORCHIDS'].shift(-500) - df['ORCHIDS']) / df['ORCHIDS']
    return df

df_constipated = calc_basket_orchids_return(df_constipated, 500, -5.2917)
df_constipated = calc_basket_orchids_return(df_constipated, 400, -4.93117)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1)

# Plot for 500its
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['basket_return_500its'], mode='lines', name='Basket Return 500its', line=dict(color='blue')), row=1, col=1)
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['expected_orchids_return_500its'], mode='lines', name='Expected Orchids Return 500its', line=dict(color='red')), row=1, col=1)
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['actual_orchids_return_500its'], mode='lines', name='Actual Orchids Return 500its', line=dict(color='green')), row=1, col=1)

# Plot for 400its
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['basket_return_400its'], mode='lines', name='Basket Return 400its', line=dict(color='blue')), row=2, col=1)
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['expected_orchids_return_400its'], mode='lines', name='Expected Orchids Return 400its', line=dict(color='red')), row=2, col=1)
fig.add_trace(go.Scatter(x=df_constipated.index, y=df_constipated['actual_orchids_return_400its'], mode='lines', name='Actual Orchids Return 400its', line=dict(color='green')), row=2, col=1)

fig.update_layout(height=600, width=1500, title_text="Orchids Return Analysis")
fig.show()
```
